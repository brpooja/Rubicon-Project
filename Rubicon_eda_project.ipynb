```python
# 1. Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load the dataset
df=pd.read_csv("C:/Users/brpoo/Downloads/udemy_courses.csv")

# 3. Show first 5 rows
print(df.head())
```

```python
# Check number of rows and columns
print("Total rows and columns:", df.shape)
```

```python
# Show column names
print("Column names:", df.columns.tolist())
```

```python
# Show data types and missing values
print(df.info())
print(df.isnull().sum())
```

```python
#1. Count of Free vs Paid Courses

print(df['is_paid'].value_counts())

# Simple bar chart
df['is_paid'].value_counts().plot(kind='bar', title='Free vs Paid Courses')
plt.xlabel("Is Paid?")
plt.ylabel("Number of Courses")
plt.show()
```

```python
#2. Number of Courses by Subject

print(df['subject'].value_counts())

df['subject'].value_counts().plot(kind='bar', title='Courses by Subject', color='skyblue')
plt.xlabel("Subject")
plt.ylabel("Number of Courses")
plt.show()
```

```python
#3. Course Levels

print(df['level'].value_counts())

df['level'].value_counts().plot(kind='bar', title='Course Levels', color='lightgreen')
plt.xlabel("Level")
plt.ylabel("Count")
plt.show()
```

```python
#4. Average Price of Paid Courses

# Filter paid courses only
paid_courses = df[df['is_paid'] == True]

# Average price
print("Average price of paid courses: ₹", paid_courses['price'].mean())
```

```python
# 5. Top 5 Most Popular Courses (by subscribers)

top_courses = df.sort_values('num_subscribers', ascending=False).head(5)
print(top_courses[['course_title', 'num_subscribers']])
```
